```python
import numpy as np
dis_airplane = np.sum(df.loc[df['mode'] == 'airplane']['distance'].values)
dis_bike = np.sum(df.loc[df['mode'] == 'bike']['distance'].values)
dis_boat = np.sum(df.loc[df['mode'] == 'boat']['distance'].values)
dis_bus = np.sum(df.loc[df['mode'] == 'bus']['distance'].values)
dis_car = np.sum(df.loc[df['mode'] == 'car']['distance'].values)
dis_motorcycle = np.sum(df.loc[df['mode'] == 'motorcycle']['distance'].values)
dis_run = np.sum(df.loc[df['mode'] == 'run']['distance'].values)
dis_subway = np.sum(df.loc[df['mode'] == 'subway']['distance'].values)
dis_taxi = np.sum(df.loc[df['mode'] == 'taxi']['distance'].values)
dis_train = np.sum(df.loc[df['mode'] == 'train']['distance'].values)
dis_walk = np.sum(df.loc[df['mode'] == 'walk']['distance'].values)
```

```python
labels = ['airplane', 'bike', 'boat', 'bus', 'car', 'motorcycle', 'run', 'subway', 'taxi','train', 'walk']
dis = (dis_airplane/1000, dis_bike/1000, dis_boat/1000, dis_bus/1000, dis_car/1000,dis_motorcycle/1000,dis_run/1000,dis_subway/1000,dis_taxi/1000,dis_train/1000,dis_walk/1000)
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(15, 10))
rects1 = ax.bar(x, dis, width, label='')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Distance/km', fontsize=20)
ax.set_title('Distance per mode', fontsize=26)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=16)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{0:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)


fig.tight_layout()

plt.show()
```

```python
from datetime import datetime
df['datetime'] = list(map(lambda s: datetime.strptime(s, '%Y-%m-%d %H:%M:%S'), df['datetime'].values))
```

```python
import sys
```

```python
mode_set = set(df['mode'].values)
num_tracks = {}

for item in mode_set:
    num_track = 1
    df_tmp = df.loc[df['mode'] == item]
    df_tmp.index = [i for i in range(len(df_tmp))]
    print('test')
    for i, row in df_tmp.iterrows():
        sys.stdout.write('\r{} / {}'.format(i, len(df_tmp)))
        if i == 0: continue
        last_row = df.iloc[i - 1]
        duration = row['datetime'] - last_row['datetime']
        if duration.seconds > 300:
            num_track += 1
    print('number of track of {}: {}'.format(item, num_track))
    num_tracks[item] = num_track
```

```python
labels = mode_set
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(15, 10))
rects1 = ax.bar(x, list(num_tracks.values()), width, label='')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Numbers of tracks', fontsize=20)
ax.set_title('Numbers of tracks per mode', fontsize=26)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=16)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)


fig.tight_layout()

plt.show()
```

```python
mode_set = set(df['mode'].values)
ave_dis = {}

for item in mode_set:
    num_of_track = 1
    dis_track = [0]*num_tracks[item]
    df_tmp = df.loc[df['mode'] == item]
    df_tmp.index = [i for i in range(len(df_tmp))]
    print('test')
    for i, row in df_tmp.iterrows():
        sys.stdout.write('\r{} / {}'.format(i, len(df_tmp)))
        if i == 0: continue
        last_row = df.iloc[i - 1]
        duration = row['datetime'] - last_row['datetime']
        if duration.seconds > 300:
            num_of_track += 1
            dis_track[num_of_track-1] += df.iloc[i].distance
        else:
            dis_track[num_of_track-1] += df.iloc[i].distance
    ave_dis[item]=np.sum(dis_track)/num_tracks[item]
    print('Average distance per track of {}: {}'.format(item, ave_dis[item]))
```

```python
labels = mode_set
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(15, 10))
rects1 = ax.bar(x, list(ave_dis.values()), width, label='')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Average distance per track/m', fontsize=20)
ax.set_title('Average distance per track per mode', fontsize=26)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=16)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{0:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)


fig.tight_layout()

plt.show()
```
